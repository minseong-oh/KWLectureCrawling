# 토큰화 진행해서 평점별로 특징 찾기

### 사전에 단어 추가 가능있음

```python
from konlpy.tag import Okt
okt=Okt()
okt.tagset
```

```python
df
```

```python
def tokenizer(reviews):
    # 명사, 부사, 형용사만 추출
    token = dict()
    token['Noun'] = dict()        # 명사
    token['Adverb'] = dict()      # 부사
    token['Adjective'] = dict()   # 형용사
    token['Verb'] = dict()        # 동사

    for review in reviews:
        try:
            for data in okt.pos(review):
                word = data[0]
                pos = data[1]
                if pos not in token:
                    continue
                if word not in token[pos]:
                    token[pos][word] = 1
                else:
                    token[pos][word] += 1
        except:
            continue
    return token
```

```python
scope1 = tokenizer(df.star1)
scope2 = tokenizer(df.star2)
scope3 = tokenizer(df.star3)
scope4 = tokenizer(df.star4)
scope5 = tokenizer(df.star5)
```

```python
scope1
```

```python
def revSort(token):
    noun = sorted(token['Noun'].items(), key=lambda x: x[1], reverse=True)
    adj = sorted(token['Adjective'].items(), key=lambda x: x[1], reverse=True)
    adv = sorted(token['Adverb'].items(), key=lambda x: x[1], reverse=True)
    verb = sorted(token['Verb'].items(), key=lambda x: x[1], reverse=True)
    return noun, adj, adv, verb
```

```python
noun1, adj1, adv1, verb1 = revSort(scope1)
noun2, adj2, adv2, verb2 = revSort(scope2)
noun3, adj3, adv3, verb3 = revSort(scope3)
noun4, adj4, adv4, verb4 = revSort(scope4)
noun5, adj5, adv5, verb5 = revSort(scope5)
```
