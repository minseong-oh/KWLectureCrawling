```python
data = set()
duplicated = []
for i in range(df2.shape[0]):
    if (df2['과목명'][i], df2['담당교수'][i]) in data:
        duplicated.append((df2['과목명'][i], df2['담당교수'][i]))
    else:
        data.add((df2['과목명'][i], df2['담당교수'][i]))
        
```

# 각 강의평에 대한 세부정보(별점, 과제, 조모임 등)
그러기 위해서 df와 df2를 병합

```python
lectures = []
professors = []

for i in range(df.shape[0]):
    data = df['info'][i]
    data = data.split(',')
    lectures.append(data[0][2:-1])
    professors.append(data[1][2:-2])
```

```python
print(len(lectures))
print(len(professors))
```

```python
df['lecture'] = lectures
df['professor'] = professors
del df['info']
```

```python
from konlpy.tag import Okt
okt=Okt()
```

```python
okt.tagset
```

```python
# 명사, 부사, 관형사, 감탄사, 형용사만 추출해봄
token = dict()
token['Noun'] = dict()
token['Adverb'] = dict()
token['Adjective'] = dict()
token['Determiner'] = dict()
token['Exclamation'] = dict()
token['KoreanParticle'] = dict()

for text in df.review:
    try:
        for data in okt.pos(text):
            word = data[0]
            pos = data[1]
            if pos not in token:
                continue
            if word not in token[pos]:
                token[pos][word] = 1
            else:
                token[pos][word] += 1
    except:
        continue
print(token)
```
